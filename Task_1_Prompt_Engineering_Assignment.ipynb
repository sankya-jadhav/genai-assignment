## Task 1: Basic Prompt Template
In this task, I tried to replace a hardcoded string with a real LangChain PromptTemplate. I put the template outside the function so it can be reused.

```python
# import stuff
from langchain.prompts import PromptTemplate

# define global template
explain_prompt_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# function to make the prompt
def generate_explain_prompt(topic: str) -> str:
    # return formatted template
    return explain_prompt_template.format(topic=topic)

print("Task 1 Output:")
print(generate_explain_prompt("Machine Learning"))
```

## Task 2: Adding More Inputs
Now I made a template that takes multiple variables like topic, audience, and tone.

```python
# global template with 3 inputs
multi_prompt_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# generate the prompt
def generate_multi_input_prompt(topic: str, audience: str, tone: str) -> str:
    return multi_prompt_template.format(topic=topic, audience=audience, tone=tone)

# print test cases
print("--- Test Case 1 ---")
print(generate_multi_input_prompt(topic="AI", audience="beginners", tone="friendly"))

print("\n--- Test Case 2 ---")
print(generate_multi_input_prompt(topic="Python", audience="kids", tone="fun"))

print("\n--- Test Case 3 ---")
print(generate_multi_input_prompt(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Making a Variations Engine
Here I created three different templates for different styles. I used a dictionary to store them so I can just pick the right one without doing a bunch of if-else checks.

```python
# dict to hold our templates
variation_templates = {
    "Teaching": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} clearly step by step"
    ),
    "Interview": PromptTemplate(
        input_variables=["topic"],
        template="Ask 3 questions about {topic}"
    ),
    "Storytelling": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as a story"
    )
}

# function to get variation
def generate_variation(style: str, topic: str) -> str:
    # grab template from dictionary
    selected_template = variation_templates.get(style)
    if not selected_template:
        raise ValueError(f"Unknown style: {style}")

    return selected_template.format(topic=topic)

# testing with machine learning
topic_input = "Machine Learning"
print("Task 3 Outputs for topic:", topic_input)

# loop through styles and print
for style in variation_templates:
    print(f"\n[{style}]:")
    print(generate_variation(style, topic_input))
```

## Task 4: Chat Templates with Roles
In this step I used ChatPromptTemplate. I made a dictionary for different roles so the system behavior changes based on the role passed in.

```python
# import chat template
from langchain.prompts import ChatPromptTemplate

# map roles to system instructions
role_behaviors = {
    "teacher": "You are an expert teacher sharing knowledge simply and effectively.",
    "interviewer": "You are a technical interviewer assessing a candidate's skills.",
    "motivator": "You are an enthusiastic life coach building confidence."
}

# chat template setup
base_chat_template = ChatPromptTemplate.from_messages([
    ("system", "{behavior}"),
    ("user", "Here is the topic: {topic}")
])

# generator function
def generate_chat_prompt(role: str, topic: str) -> str:
    # get behavior for the role
    behavior = role_behaviors.get(role)
    if not behavior:
        raise ValueError(f"Unknown role: {role}")

    messages = base_chat_template.format_messages(behavior=behavior, topic=topic)
    return "\n".join(f"{msg.type.capitalize()}: {msg.content}" for msg in messages)

# test output for different roles
roles = ['teacher', 'interviewer', 'motivator']
topic_input = "Neural Networks"

for role in roles:
    print(f"\n--- Role: {role.capitalize()} ---")
    print(generate_chat_prompt(role, topic_input))
# simple check for valid inputs
```

## Task 5: Doing Input Validation
Just adding a basic validation function to make sure someone doesn't pass in a weird audience or tone.

```python
# validation logic
def validate_inputs(audience: str, tone: str) -> None:
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # raise error if it doesn't match
    if audience not in valid_audiences:
        raise ValueError(f"Invalid audience '{audience}'. Must be one of: {valid_audiences}")
    if tone not in valid_tones:
        raise ValueError(f"Invalid tone '{tone}'. Must be one of: {valid_tones}")
    print(f"Validation successful: Audience='{audience}', Tone='{tone}'")

# run a pass and fail case
print("--- Valid Case ---")
validate_inputs(audience="beginner", tone="fun")

print("\n--- Invalid Case ---")
try:
    validate_inputs(audience="kids", tone="angry")
except ValueError as e:
  print(e)
```

## Task 6: The Prompt Generator
This combines the validation from earlier with a dictionary of templates to act like a small prompt generator app.

```python
# templates for the app
app_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Ask {audience} 3 {tone} questions about {topic} for an interview"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    )
}

# main app function
def generate_prompt_app(topic: str, audience: str, tone: str, style: str) -> str:
    # check inputs first
    validate_inputs(audience=audience, tone=tone)

    # get template
    selected_template = app_templates.get(style)
    if not selected_template:
        raise ValueError(f"Invalid style. Must be one of: {list(app_templates.keys())}")

    # fill in template
    return selected_template.format(topic=topic, audience=audience, tone=tone)

# print app result
print("\nTask 6 Output:")
print(generate_prompt_app(topic="Neural Networks", audience="beginner", tone="fun", style="storytelling"))
```

## Task 7: Reusability Test
Finally, I just made one generic template and passed 5 different sets of inputs to it to show how reusable it is.

```python
# one big reusable template
reusable_generic_template = PromptTemplate(
    input_variables=["topic", "audience", "tone", "style"],
    template="Explain {topic} specifically for {audience} with a {tone} tone. Keep the output format in a {style} style."
)

# test inputs
inputs = [
    {"topic": "Data Science", "audience": "beginner", "tone": "casual", "style": "teaching"},
    {"topic": "Quantum Mechanics", "audience": "expert", "tone": "formal", "style": "teaching"},
    {"topic": "Python OOP", "audience": "intermediate", "tone": "fun", "style": "storytelling"},
    {"topic": "SQL Joins", "audience": "beginner", "tone": "casual", "style": "interview"},
    {"topic": "Docker", "audience": "intermediate", "tone": "formal", "style": "teaching"}
]

# loop and run
print("Task 7 Outputs:")
for i, params in enumerate(inputs, 1):
    print(f"\nRun {i}:")
    print(reusable_generic_template.format(**params))
```
