### 2. Basic LLM Call
*A simple call to Groq's extremely fast LLaMA 3.3 model.*

```python
import os
from langchain_groq import ChatGroq

# Set your Groq API key here
os.environ["GROQ_API_KEY"] = "gsk_R...."

# Initialize the Groq LLM
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)

# Send a basic message to the model
response = llm.invoke("Explain LangChain in simple terms")
print("AI Response:\n", response.content)
```

### 3. PromptTemplate Usage
*Creating dynamic, reusable prompts with variables.*

```python
from langchain_core.prompts import PromptTemplate

# Define the template with a placeholder
template = "Tell me a short fact about {topic}."
prompt = PromptTemplate(input_variables=["topic"], template=template)

# Fill in the placeholder
formatted_prompt = prompt.format(topic="Space")
print("Formatted Prompt:", formatted_prompt)
```

### 4. Simple Chain
*Connecting the prompt and the LLM seamlessly.*

```python
from langchain.chains import LLMChain

# Create a chain that glues the prompt and the LLM together
chain = LLMChain(llm=llm, prompt=prompt)

# Run the chain directly with the variable
result = chain.invoke({"topic": "Artificial Intelligence"})
print("Chain Output:", result['text'])
```

### 5. Memory Example
*Giving the LLM context of past conversations.*

```python
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory

# Set up the memory to remember chat history
memory = ConversationBufferMemory()

# Create a conversation chain using our Groq LLM
conversation = ConversationChain(llm=llm, memory=memory)

# Talk to the AI
print("Interaction 1:", conversation.predict(input="Hi, my name is Alex."))
print("Interaction 2:", conversation.predict(input="Can you remember my name?"))
```

### 6. Agent with Tool
*Allowing the LLM to use an external tool (Math) to solve problems.*

```python
from langchain.agents import load_tools, initialize_agent, AgentType

# Load a built-in math calculator tool
tools = load_tools(["llm-math"], llm=llm)

# Initialize the agent
agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True # Set to True to see the agent's internal thinking process
)

try:
    # Ask it a math problem
    result = agent.invoke({"input": "What is 15 multiplied by 4, and then divided by 2?"})
    print("Agent Result:", result['output'])
except Exception as e:
    print("Agent Error:", e)
```

### 7. Document Loaders & Indexes (RAG)
*Loading a text file and querying it explicitly using open-source HuggingFace Embeddings (free).*

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

# 1. Create file
with open("sample.txt", "w") as file:
    file.write("LangChain is an AI framework built by Harrison Chase in 2022.")

# 2. Load document
loader = TextLoader("sample.txt")
documents = loader.load()

# 3. Embeddings
embedding = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# 4. Create vector DB
db = FAISS.from_documents(documents, embedding)

# 5. Query
docs = db.similarity_search("Who created LangChain?")

print("RAG Answer:", docs[0].page_content)
```
